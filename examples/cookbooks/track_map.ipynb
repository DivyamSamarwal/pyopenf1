# Drawing a Track Map with pyopenf1

In this cookbook, we'll use `pyopenf1` to fetch location telemetry for a specific driver and draw a track map using `matplotlib` and `pandas`.

```python
import matplotlib.pyplot as plt

from pyopenf1 import AsyncOpenF1Client
from pyopenf1.ext.pandas import to_dataframe


async def fetch_telemetry():
    async with AsyncOpenF1Client() as client:
        # Fetch location data for driver 1 (Max Verstappen) in a specific session
        location_data = await client.telemetry.get_location(driver_number=1, session_key=9158)
        return to_dataframe(location_data)

# Run the async function to get our dataframe
df = await fetch_telemetry()
```

Now that we have the data in a pandas DataFrame, we can easily plot the `x` and `y` coordinates to see the track.

```python
# Create the plot
plt.figure(figsize=(10, 10))

# OpenF1 location data uses x and y coordinates.
# Note: sometimes the y-axis needs to be inverted depending on the track orientation.
plt.plot(df['x'], df['y'], color='blue', linewidth=2)

plt.title("Track Map - Driver 1")
plt.axis('equal') # Ensure the track aspect ratio is correct
plt.axis('off')   # Hide axes for a cleaner look
plt.show()
```
